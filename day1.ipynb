## 1: Thinking Algorithmically

The biggest challenge in programming isn't learning the syntax, it's learning
how to **break a problem down** into small, concrete steps that a computer can
follow.

A computer is incredibly fast but incredibly literal. It will do *exactly* what
you tell it to. So before writing code, you
need to figure out: **what ***exactly*** are the steps?**

This is called **algorithmic thinking**, and the tool we use to practice it is
**pseudocode**: writing out your plan in plain English (or whatever language
you think in) before translating it to Python.

### Example: The PB&J problem

Imagine you need to write instructions for a robot to make a peanut butter and
jelly sandwich. The robot follows instructions *literally*. If you say
"put peanut butter on the bread," it might place the entire jar on top of a
loaf. You need to be specific:

```
1. Take two slices of bread from the bag
2. Open the peanut butter jar
3. Pick up a knife
4. Scoop peanut butter with the knife
5. Spread peanut butter on one side of slice 1
6. Open the jelly jar
7. Scoop jelly with the knife
8. Spread jelly on one side of slice 2

```

That's pseudocode! It's not Python, but it's specific enough that you could
translate each step into code if you needed to.

In this notebook, I want you to focus on the 4 key steps to approaching any problem:

1. **Understand** the problem — what are the inputs? What's the desired output?
2. **Plan** in pseudocode — write out the steps in English
3. **Translate** to Python — turn each pseudocode line into code
4. **Test** — run it and check if the output is what you expected

## 2: `for` Loops — The Basics

A **loop** lets you repeat a block of code multiple times. The most common type
is the `for` loop.

### Looping over a list

The simplest use of a `for` loop: do something **for each item** in a collection.

````python
organisms = ['E. coli', 'yeast', 'zebrafish', 'mouse', 'human']

for organism in organisms:
    print(f"I study {organism}!")
````

Let's trace through what happens step by step:

| Iteration | `organism` | Output |
|-----------|------------|--------|
| 1st | `'E. coli'` | `I study E. coli!` |
| 2nd | `'yeast'` | `I study yeast!` |
| 3rd | `'zebrafish'` | `I study zebrafish!` |
| 4th | `'mouse'` | `I study mouse!` |
| 5th | `'human'` | `I study human!` |

Each time through the loop, the variable `organism` takes on the **next value**
from the list. When the list runs out, the loop stops.

### Looping with `range()`

Sometimes you just want to repeat something a certain
number of times. That's what `range()` is for.

````python
# range(5) gives you the numbers [0, 1, 2, 3, 4] as a LIST. This is what lets us 
# *iterate* through them!
for i in range(5):
    print(f"Step {i}")
````

````python
# range is a function that takes `(start, stop)`. Just like slicing, stop is NOT included
for i in range(2, 7):
    print(i)
````

You can loop over any *iterable*. Think of iterables of objects that are composed of a sequence of other objects. For example, what is a string but a sequence of characters? This means you can loop directly over the characters of a string:

````python
dna = "ATGCGA"

for base in dna:
    print(base)
````

## 4: `while` Loops

A `for` loop is great when you know **how many times** to repeat. But sometimes
you want to keep going **until something happens**. That's a `while` loop.

```python
while condition:
    # keep doing this as long as condition is True
```

**Be careful:** if the condition never becomes `False`, the loop runs forever!


````python
# Example: population growth
# A colony of bacteria doubles every generation.
# How many generations until the population exceeds 1000?

population = 1
generations = 0

while population <= 1000:
    population = population * 2
    generations = generations + 1

print(f"After {generations} generations, population = {population}")
````

Let's trace through the first few iterations:

| Generation | Population | `population <= 1000`? |
|------------|------------|----------------------|
| 0 | 1 | True, keep going |
| 1 | 2 | True, keep going |
| 2 | 4 | True, keep going |
| ... | ... | ... |
| 10 | 1024 | False, stop! |

**When to use which:**
- **`for`**: you know how many times ("do this 100 times", "for each item in this list")
- **`while`**: you don't know how many times ("keep going until X happens")

### The Tracker

Keep track of the "best so far" — the max, min, first match, etc.

````python
# Tracker: find the smallest value
import math
temps = [36.2, 37.8, 35.1, 36.5, 38.4]
smallest = math.inf        # What do we start the tracker at? Why not infinity!
for t in temps:
    if t < smallest:
        smallest = t       # found a new smallest!
print(smallest)        
````

### The List Builder

Start with an empty list and `.append()` to it at each step. This one is **really important** and you have likely seen/will see this in many problems.

````python
# List builder: collect the squares of 0 through 9
squares = []               # start with empty list
for i in range(10):
    squares.append(i ** 2) # append new value
print(squares)
````

## 6: Looping with Indices and Nested Loops

### Looping by index

We saw that when given a list of items, we can loop through all the items and do stuff with them. Sometimes you need to know *where* you are in a list, not just what the current
item is. Remember how we used range to do stuff a certain amount of times? Well since we can use `len` to get a **number** for the **length of the list**, we can combine range and len to loop over the indices using `range(len(...))`:

````python
genes = ['TP53', 'BRCA1', 'MYC', 'EGFR']

# Looping by value (when you don't need the index)
for gene in genes:
    print(gene)


# Looping by index (when you DO need the index)
for i in range(len(genes)):
    print(f"Gene {i}: {genes[i]}")
````

**When do you need the index?** Whenever you need to:
- Know the position of an item
- Access elements from **two lists at the same time**
- Modify a list in place

## 7: Functions

You might have noticed that we keep writing the same GC content calculation
over and over. Wouldn't it be nice to write it once and reuse it?

That's exactly what a **function** is — a reusable block of code with a name.

```python
def function_name(parameter1, parameter2):
    # do something
    return result
```

````python
# A simple function
def say_hello(name):
    return f"Hello, {name}!"

# Call it:
message = say_hello("LS50")
print(message)
````

````python
# A more useful function: convert Fahrenheit to Celsius
def fahrenheit_to_celsius(f):
    c = (5 / 9) * (f - 32)
    return c

print(fahrenheit_to_celsius(98.6))  
print(fahrenheit_to_celsius(212))   
print(fahrenheit_to_celsius(32))    
````

The key idea: `return`  *returns* a value **back** to whoever called the function.
You can then store that value in a variable, print it, or use it in other
calculations.
