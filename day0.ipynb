## 2: Indexing and Slicing

Now that we know about lists and strings, we need a way to access individual
elements or chunks of them. This is called **indexing** and **slicing**.

**Important:** Python counts starting from **0**, not 1!

````python
dna = "ATGCGATCGA"

# Indexing: get a single element
print(dna[0])   # first character
print(dna[3])   # fourth character
print(dna[-1])  # last character
````

````python
# Slicing: get a range of elements using [start:stop]
# Note: the stop index is NOT included!
print(dna[0:3])   # first three characters
print(dna[4:7])   # characters at index 4, 5, 6
print(dna[:5])    # from the beginning to index 5 (not included)
print(dna[5:])    # from index 5 to the end
````

Indexing works the same way for lists:

````python
organisms = ['E. coli', 'yeast', 'zebrafish', 'mouse', 'human']

print(organisms[0])    # first element
print(organisms[-1])   # last element
print(organisms[1:3])  # a slice: ['yeast', 'zebrafish']
````

## 3: Operations

### Quick review: math operators

You're probably already comfortable with the basic math operators. Here's a
quick reference:

| Operator | Meaning | Example |
|----------|---------|---------|
| `+` | addition | `3 + 2` → `5` |
| `-` | subtraction | `3 - 2` → `1` |
| `*` | multiplication | `3 * 2` → `6` |
| `/` | division | `7 / 2` → `3.5` |
| `**` | exponent | `3 ** 2` → `9` |
| `%` | modulo (remainder) | `7 % 2` → `1` |
| `//` | integer division | `7 // 2` → `3` |

### String operations

Strings have their own set of operations. Some of them look like math, but they
do something different!

````python
# Concatenation: glue strings together with +
greeting = "Hello" + " " + "LS50!"
print(greeting)

# Repetition: repeat a string with *
print("AT" * 5)

# Length: how many characters?
dna = "ATGCGATCGA"
print(len(dna))
````

Strings also come with built-in **methods** — little tools attached to every
string. You call them with a dot: `my_string.method()`.

````python
dna = "ATGCGATCGA"

# .count() -- count occurrences of a substring
print("Number of G's:", dna.count('G'))

# .replace() -- swap one substring for another
rna = dna.replace('T', 'U')
print("RNA:", rna)

# .lower() and .upper()
gene = "Brca1"
print(gene.upper())
print(gene.lower())

# .split() -- break a string into a list
data = "gene1,gene2,gene3,gene4"
gene_list = data.split(',')
print(gene_list)
````

### f-strings

One last string trick that will come in very handy: **f-strings** let you embed
variables directly inside a string. Just put an `f` before the quotes and wrap
your variables in curly braces `{}`.

````python
species = "E. coli"
growth_rate = 0.4

# Without f-string (clunky)
print("The growth rate of " + species + " is " + str(growth_rate) + " per hour.")

# With f-string (much cleaner!)
print(f"The growth rate of {species} is {growth_rate} per hour.")
````

## 4: Booleans and Comparisons

A **boolean** is a value that is either `True` or `False`. You've actually
already seen booleans — remember the `type()` exercise at the top of this
notebook?

Booleans are the foundation of **logic** in programming. We create them using
**comparison operators**:

````python
x = 10

print(x == 10)   # equal to
print(x != 5)    # not equal to
print(x > 7)     # greater than
print(x < 3)     # less than
print(x >= 10)   # greater than or equal to
print(x <= 9)    # less than or equal to
````

You can combine booleans using `and`, `or`, and `not`:

````python
temperature = 37
ph = 7.4

# `and` -- both conditions must be True
print(temperature == 37 and ph > 7.0)   # True

# `or` -- at least one condition must be True
print(temperature > 40 or ph > 7.0)     # True (pH is > 7)

# `not` -- flips True to False and vice versa
print(not temperature == 37)             # False
````

You can also check membership in a string or list using the `in` keyword:

````python
print('ATG' in 'ATGCGATCGA')            # True -- 'ATG' is a substring
print('mouse' in ['rat', 'mouse', 'human'])  # True -- 'mouse' is in the list
print('X' in 'ATGC')                    # False
````

## 5: Conditionals

Now that we can ask True/False questions, we can make our code **do different
things** depending on the answer. This is called **conditional logic**, and it
uses `if`, `elif` (short for "else if"), and `else`.

The basic structure looks like this:

```python
if condition:
    # do this if condition is True
elif other_condition:
    # do this if the first was False but this one is True
else:
    # do this if none of the above were True
```

**Important:** notice the colon `:` at the end of each line, and the
**indentation** (4 spaces) for the code inside each block. Python uses
indentation to know which code belongs to which block!

````python
# Example: classify a pH value
ph = 7.4

if ph < 7:
    print("Acidic")
elif ph == 7:
    print("Neutral")
else:
    print("Basic")
````

Try changing the value of `ph` above and re-running the cell. Can you make it
print each of the three messages?

You can chain as many `elif`s as you need:

````python
# Example: classify body temperature (in Celsius)
temp = 38.5

if temp < 35:
    print("Hypothermia")
elif temp < 37.5:
    print("Normal")
elif temp < 39:
    print("Low-grade fever")
else:
    print("High fever")
````

## 6: User Input

So far, we've been hard-coding all our values. But what if we want the person
running the code to provide the value? That's what `input()` is for!

When Python hits an `input()` call, it **pauses** and waits for you to type
something. Whatever you type gets returned as a **string**.

````python
# Run this cell and type your name when prompted!
name = input("What is your name? ")
print(f"Hello, {name}! Welcome to LS50 coding camp.")
````

**Watch out!** `input()` always returns a **string**, even if the user types a
number. If you want to do math with it, you need to **cast** it using `int()` or
`float()`:

````python
# See the problem:
age_string = input("How old are you? ")
print(type(age_string))  # It's a string!

# Fix it by converting to int:
age = int(age_string)
print(type(age))          # Now it's an int!
print(f"In 10 years you'll be {age + 10}.")
````
